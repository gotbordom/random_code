```python
%matplotlib inline
from matplotlib import pyplot
pyplot.style.use('ggplot')
import numpy
numpy.seterr(divide='raise') # Raise exception on divide by zero

class FDict(dict):
    "Collection of labeled functions"
    def add(self, name):
        def _add(f):
            self[name] = f
        return _add

tests = FDict()

@tests.add('x^2-2')
def f(x):
    return x**2 - 2, 2*x

@tests.add('cos(x)-x')
def f(x):
    return numpy.cos(x) - x, -numpy.sin(x) - 1

def plotf(f, label=None):
    x = numpy.linspace(-3, 3)
    return pyplot.plot(x, f(x)[0], label=label)
    
for name, f in tests.items():
    plotf(f, name)
    pyplot.legend()
```

```python
def bisect(f, a, b, tol=1e-10):
    "Iterative (rather than recursive) bisection method"
    fa, fb = f(a)[0], f(b)[0]
    if fa*fb > 0:
        raise ValueError('Interval function does not change sign, may not contain root')
    history = []
    while b-a > tol:
        mid = (a + b)/2.
        fmid = f(mid)[0]
        history.append((mid, fmid))
        if fa*fmid <= 0:
            b, fb = mid, fmid
        else:
            a, fa = mid, fmid
    return numpy.array(history)

def newton(f, a, b, tol=1e-10):
    """Newton method without a line search
    """
    x = (a + b) / 2
    history = []
    for i in range(100):
        fx, dfx = f(x)
        history.append((x, fx))
        if numpy.abs(fx) < tol:
            break
        try:
            x -= fx / dfx
        except FloatingPointError:
            break
    return numpy.array(history)

def plot_conv(tests, methods):
    """Plot the convergence of a group of methods on a dict of test problems"""
    fig = pyplot.figure()
    axerror = pyplot.subplot(2,1,1)
    axresid = pyplot.subplot(2,1,2)
    for name, f in tests.items():
        for method in methods:
            h = method(f, -1, 3)
            label = '{} {}'.format(method.__name__, name)
            axerror.semilogy(numpy.abs(h[:-1,0] - h[-1,0]), label=label)
            axresid.semilogy(numpy.abs(h[:,1]), label=label)
    axerror.set_ylabel('Error $|x_i - x_*|$')
    axresid.set_ylabel('Residual $|f(x_i)|$')
    axerror.set_ylim(1e-9, 10)
    axresid.set_ylim(1e-9, 10)
    axresid.set_xlabel('Iteration $i$')
    fig.legend();
    
plot_conv(tests, (bisect, newton))
```

```python
def ismonotonic(f):
    x = numpy.linspace(-5, 5)
    fx = f(x)[0]
    return numpy.all(fx[1:] > fx[:-1])

def isdivergent(f):
    h = newton(f, -1, 3)
    hx = h[:,0]
    return numpy.abs(hx[-1]) > numpy.abs(hx[-2])

assert numpy.abs(tests['monotonic_diverge'](0)[0]) < 1e-10
assert ismonotonic(tests['monotonic_diverge'])
assert isdivergent(tests['monotonic_diverge'])
```

## Secant method

The secant method is similar to Newton's method, but instead of evaluating $f'(x)$ directly, it uses the approximation
$$ f'(x_i) = \frac{f(x_{i} - f(x_{i-1})}{x_i - x_{i-1}} . $$

```python
def secant(f, a, b, tol=1e-10):
    "Solve f(x) = 0 using the secant method"
    history = []
    xlast, flast = b, f(b)[0] # We'll use x=b as our last evaluation to initialize
    x = (a + b) / 2           # And start with the midpoint as the current guess
    for i in range(100):
        fx = f(x)[0]
        history.append((x, fx))
        if numpy.abs(fx) < tol:
            break
        # YOUR CODE HERE
        try:
            dfx = (fx-flast)/(x-xlast)
            xlast = x
            flast = fx
            x -= fx / dfx
        except FloatingPointError:
            break
            
        #raise NotImplementedError()
    return numpy.array(history)

plot_conv(tests, (newton, secant))
```

```python
assert numpy.allclose(secant(tests['x^2-2'], -1, 3)[-1],
                      [numpy.sqrt(2), 0])
```
